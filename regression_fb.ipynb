```python
#import necessary libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
fb = pd.read_csv("D:/self study/regression/uci-facebook-metrics/fb_clean_eda.csv", index_col = 0)
fb.head()
```

```python
#dropping share, comment and like columns
fb = fb.drop(["comment","share","like"], axis = 1)
```

```python
categorical_var = fb[["Category","Type","Paid"]]
numerical_var = fb.drop(["Category","Type","Paid","Total Interactions"], axis = 1)
cat_dummies = pd.get_dummies(categorical_var)
```

```python
cat_dummies.head()
```

```python
x = pd.concat([numerical_var, cat_dummies], axis = 1, sort = False)
y = fb[["Total Interactions"]]
```

```python
# implementiong multiple linear regression model 
fb.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import cross_val_score
```

```python
#define standard scalar
#and transform data
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
mm_x = mm.fit_transform(x)
mm_y = mm.fit_transform(y)
```

```python
#buiding and training model
model = LinearRegression()
fold_ = RepeatedKFold(n_splits = 10, n_repeats = 5, random_state = 1)
score = cross_val_score(model, mm_x, mm_y, cv = fold_)
print(score.mean())
```
